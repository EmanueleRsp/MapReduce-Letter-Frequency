```python

# Creare un DataFrame vuoto
df = pd.DataFrame()

# Navigare attraverso le sottocartelle e i file
for root, dirs, files in os.walk("../resources/normal/output/output_00055"):
    for file in files:
        if file == "frequency.txt":
            # Leggere i dati nel DataFrame
            temp_df = pd.read_csv(os.path.join(root, file), names=['letter', 'frequency'], sep='\t')
            # Aggiungere una colonna per il linguaggio
            temp_df['language'] = os.path.basename(root)
            # Concatenare questo DataFrame con il DataFrame principale
            df = pd.concat([df, temp_df])

# Riempire i valori mancanti con 0
df = df.fillna(0)

# Ordinare il DataFrame in base alla lettera
df = df.sort_values('letter')

# Togli "bible-" da ogni valore di language 
df['language'] = df['language'].str.replace('bible-', '').str.capitalize()
languages = df['language'].unique()

# Calcola il numero di righe per i subplot
rows = int(np.ceil(len(languages) / 2))

# Creare un grafico a barre per ogni lingua
fig, axs = plt.subplots(rows, 2, figsize=(15, rows*5))

# Rendi axs sempre 2D per evitare problemi quando c'è solo una riga
if rows == 1:
    axs = np.array([axs])

for i, language in enumerate(languages):
    ax = axs[i//2, i%2]
    df_lang = df[df['language'] == language]
    df_lang = df_lang.sort_values('frequency', ascending=False)
    ax.bar(df_lang['letter'], df_lang['frequency'])
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    ax.set_title(language)

# Rimuovi gli assi vuoti se il numero di lingue non è un multiplo di 2
if len(languages) % 2:
    fig.delaxes(axs[-1, -1])

plt.tight_layout()
plt.savefig('png/languages.png')
plt.show()
```

```python
def divide_in_groups(df, n):
    df = df.sort_values(by=['letter', 'language'])
    letters = sorted(df['letter'].unique())
    groups = [letters[i::n] for i in range(n)]

    fig, axs = plt.subplots(n, 1, figsize=(12, n * 6))

    # Se n è 1, convertire axs in un array di un elemento
    if n == 1:
        axs = np.array([axs])

    # Creare un grafico a barre per ogni gruppo di lettere
    for ax, letts in zip(axs, groups):
        df_subset = df[df['letter'].isin(letts)]
        sns.barplot(x='letter', y='frequency', hue='language', data=df_subset, palette='viridis', ax=ax)
        ax.grid(axis='y', linestyle='--', alpha=0.7)

    plt.tight_layout()
    plt.savefig('png/frequencies.png')
    plt.show()

# Chiamare la funzione con il numero desiderato di gruppi
divide_in_groups(df, 2)
```
